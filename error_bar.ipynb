```python
import numpy as np
import matplotlib.pyplot as plt

def plot_error_bar(n_groups_results):
    """
    绘制误差条图，横轴为 groud truth，纵轴为预测值的标准差。
    
    :param n_groups_results: n 组实验的二维数组列表，每个二维数组为 [[预测值, groud truth], ...]
    """
    # 将所有组的数据整合到一个大列表中
    all_data = np.concatenate(n_groups_results, axis=0)
    
    # 提取 groud truth 和预测值
    predictions = all_data[:, 0]  # 预测值
    groud_truths = all_data[:, 1].astype(int)  # groud truth（转换为整数）

    # 找到 groud truth 的唯一值
    unique_groud_truths = np.unique(groud_truths)

    # 创建用于存储每个 groud truth 的均值和标准差的列表
    means = []
    stds = []

    # 逐个计算每个 groud truth 的预测值均值和标准差
    for gt in unique_groud_truths:
        # 筛选出 groud truth 为 gt 的所有预测值
        values = predictions[groud_truths == gt]
        means.append(np.mean(values))
        stds.append(np.std(values))

    # 绘制误差条图
    plt.figure(figsize=(8, 6))
    plt.errorbar(unique_groud_truths, means, yerr=stds, fmt='o', capsize=5, capthick=2, ecolor='red', label='Error Bar')

    # 绘制理想的参考线（groud truth 对应的值）
    plt.plot(unique_groud_truths, unique_groud_truths, label='Ideal Line', linestyle='--', color='blue')
    
    plt.xlabel('Ground Truth')
    plt.ylabel('Predicted Values')
    plt.title('Error Bar Plot of Predicted Values with Standard Deviation')
    plt.xticks(unique_groud_truths)  # 设置x轴的刻度为 groud truth
    plt.grid(True)
    plt.legend()
    plt.show()

```

## Database: backdoor, all, linear 10

```python
l10_res0, l10_de_res0 = PoisonDetector.predict(preditc_dir='Sequences/bkd-all-linear10-test/', model_path='bkd_all_l10_model.pth', graph='linear', graph_size=10)
l10_res1, l10_de_res1 = PoisonDetector.predict(preditc_dir='Sequences/bkd-all-linear10-test1/', model_path='bkd_all_l10_model.pth', graph='linear', graph_size=10)
l10_res2, l10_de_res2 = PoisonDetector.predict(preditc_dir='Sequences/bkd-all-linear10-test2/', model_path='bkd_all_l10_model.pth', graph='linear', graph_size=10)
l10_res3, l10_de_res3 = PoisonDetector.predict(preditc_dir='Sequences/bkd-all-linear10-test3/', model_path='bkd_all_l10_model.pth', graph='linear', graph_size=10)
```

```python
n_groups_results = [l10_de_res0, l10_de_res1, l10_de_res2, l10_de_res3]
plot_error_bar(n_groups_results)
```
